```python
def getTotalPop(indicators):
    # to access both the index and row values simultaneously we use iterrows fuction
    for index, row in indicators.iterrows():
        value = row['Indicator Name']
        # check which index has the substring 'Population, total'
        if 'Population, total' in value:
            # return the value found
            return index
        
getTotalPop(indicators)
```

```python
def getGdpDollar(indicators):
    for index, row in indicators.iterrows():
        value = row['Indicator Name']
        # check which index has both the substrings 'GDP' and 'current US$' and, at the same time, does not have
        # the substring 'per capita'
        if 'GDP' in value and 'current US$' in value and 'per capita' not in value:
            # return the value found
            return index
        
getGdpDollar(indicators)
```

```python
def getGiniIndex(indicators):
    for index, row in indicators.iterrows():
        value = row['Indicator Name']
        # check which index has the substring 'Gini index'
        if 'Gini index' in value:
            # return the value found
            return index
        
getGiniIndex(indicators)
```

```python
#4

# Define the population classifications and their ranges
population_ranges = [
    (0, 1000000, 'Tiny'),
    (1000000, 5000000, 'Very Small'),
    (5000000, 15000000, 'Small'),
    (15000000, 30000000, 'Medium'),
    (30000000, 100000000, 'Large'),
    (100000000, np.inf, 'Huge')
]

#create a fuction to check the range of each population
def classify_population(population):
    for min_pop, max_pop, classification in population_ranges:
        if min_pop <= population < max_pop:
            return classification
    # if country has no population defined, return "Unknown"
    return "Unknown"

# create a copy of the dataframe so we can work and modify it without changing the original one 
filtered_df_per_capita = data.copy()  

# creating a new column with country 'GDP per capita'
filtered_df_per_capita["GDP per capita"] = data["GDP"] / data["Population"]

# create a new column for the classifications
filtered_df_per_capita['Population Classification'] = filtered_df_per_capita['Population'].apply(classify_population)

# Group the DataFrame by the new classification column and show their respective mean and std
grouped = filtered_df_per_capita.groupby('Population Classification')["GDP per capita"].agg(["mean", "std"])

print(grouped)
```

```python
#5

def create_binary_series(df, column):
    # Get mean from the selected column
    column_mean = df[column].mean()

    # Create a Series following the given rules
    binary_series = df[column].apply(lambda x: 1 if not pd.isna(x) and x > column_mean else 0 if not pd.isna(x) else np.nan)

    return binary_series

result_series = create_binary_series(data, 'GDP')

print(result_series)
```

```python
#9

# create a copy of the dataframe so we can work and modify it without changing the original one 
filtered_df_per_capita = data.copy()

# for this exercise we are dropping countries with 'hc' or 'gini' with 'nan' values, because they would be considered
# as below the average, which is not what we are looking for
filtered_df_per_capita = filtered_df_per_capita.dropna(subset=['hc', 'GINI'])

# create a new column for the classifications
filtered_df_per_capita['Population Classification'] = filtered_df_per_capita['Population'].apply(classify_population)

# create a new column with 'GDP per capita' for each country
filtered_df_per_capita["GDP per capita"] = data["GDP"] / data["Population"]

# Group the DataFrame by the classification and fetch the average GINI per group
filtered_df_per_capita['avg gini per class'] = filtered_df_per_capita.groupby('Population Classification')['GINI'].transform('mean')
# Group the DataFrame by the classification and fetch the average HC per group
filtered_df_per_capita['avg hc per class'] = filtered_df_per_capita.groupby('Population Classification')['hc'].transform('mean')

# create a new column in the df with the boolean according to hc relationship with hc mean per class
filtered_df_per_capita['hc above mean per class'] = np.where(filtered_df_per_capita['hc'] > filtered_df_per_capita['avg hc per class'], 1, 0)

# create a new column in the df with the boolean according to GINI relationship with GINI mean per class
filtered_df_per_capita['gini above mean per class'] = np.where(filtered_df_per_capita['GINI'] > filtered_df_per_capita['avg gini per class'], 1, 0)

# filtering the whole dataframe by the specified rules in the exercise
filtered_df_per_capita = filtered_df_per_capita[(filtered_df_per_capita['gini above mean per class'] == 0) & (filtered_df_per_capita['hc above mean per class'] == 0)]

# selecting the countrycode of the country with the highest 'GDP per capita' in this dataframe
max_gdp_idx = filtered_df_per_capita['GDP per capita'].idxmax()

# matching the countrycode in the pwt dataframe (since it also has the countries names), and giving the first value
# we do not want all the matching rows in the df
selected_country = pwt[pwt['countrycode'] == max_gdp_idx]['country'].head(1).values[0]

selected_country
```

```python
#10
# Reusing the function created in the previous exercises to filter the df 'getGdpDollar()', with a small change in the return
def getGdpDollarRow(indicators):
    for index, row in indicators.iterrows():
        value = row['Indicator Name']
        # check which index has both the substrings 'GDP' and 'current US$' and, at the same time, does not have
        # the substring 'per capita'
        if 'GDP' in value and 'current US$' in value and 'per capita' not in value:
            # return the value found
            return row
        
wdi_aux = wdi[wdi["Indicator Code"] == getGdpDollarRow(wdi)["Indicator Code"]] 

wdi_filtered = wdi_aux.copy()

# If a country has no GDP registered for 1980, using the first actual value of the GDP after that date
wdi_filtered['first_valid_year_after_1980'] = wdi_filtered.apply(
    lambda row: row.loc['1981':'2010'].first_valid_index() if pd.isna(row['1980']) else '1980', 
    axis=1
)

# Filling nans (if a country has no GDP registered from 1980 to 2010) with the value 0
wdi_filtered['first_valid_year_after_1980'] = wdi_filtered['first_valid_year_after_1980'].fillna(0)

# Finding the first valud GDP after 1980 by using the first valid year previously identified
wdi_filtered['first_valid_gdp_after_1980'] = wdi_filtered.apply(
    lambda row: row[row['first_valid_year_after_1980']], 
    axis=1
)

# Calculate the GDP % growth for each country, using the GDP's found above
wdi_filtered['gdp_growth'] = (
    (wdi_filtered['2010'] - wdi_filtered['first_valid_gdp_after_1980']) / 
    wdi_filtered['first_valid_gdp_after_1980']
) * 100

# Identify the country with the biggest gdp_growth (%)
max_growth_country = wdi_filtered.loc[wdi_filtered['gdp_growth'].idxmax(), 'Country Name']

print(f"The country with the largest GDP growth from the first valid year after 1980 to 2010 is: {max_growth_country}")
```
